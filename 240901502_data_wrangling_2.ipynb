```python
import pandas as pd
#Assign data
data={'Name':['Sai','Yuvasri','Poovannan',
'Sabharish','Sangamithra','Rithika','Rubasri'],
'Age':[20,18,19,20,18,17,19],
'Gender':['M','F','M','M','F','F','F'],
'Marks':[90,76,'NaN',74,65,'NaN',71]}
#convert into datafram
df=pd.DataFrame(data)
#display data
print("Original DataFrame:\n",df)

# Correct average calculation
total_marks = 0
count_numeric_marks = 0
for ele in df['Marks']:
    if str(ele).isnumeric():
        total_marks += int(ele)
        count_numeric_marks += 1

# Calculate average after iterating through all elements
if count_numeric_marks > 0:
    average_value = total_marks / count_numeric_marks
else:
    average_value = 0 # Handle case where no numeric values are found

# Convert 'Marks' column to numeric, coercing 'NaN' strings to np.nan
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')
# Replace missing values (now np.nan) with the calculated average
df['Marks'] = df['Marks'].fillna(average_value)

#Display data
print("\nReplacing NaN with average marks:\n",df)
```
